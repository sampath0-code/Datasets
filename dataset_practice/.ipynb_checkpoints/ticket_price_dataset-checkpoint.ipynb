```python
 # updating the Date of journey, dep_time, arrival time from objects to date_time
```

```python
# creating a function to change the 3 features
def change_date_time(col):
    df[col] = pd.to_datetime(df[col])
```

```python
for i in ['Date_of_Journey','Dep_Time', 'Arrival_Time'] :
    change_date_time(i)
```

```python
# feature selection and updation 
# updating the date in the date of journey into the seperate columns day, month and year for the machine understandability
```

```python
df['journey_day'] = df['Date_of_Journey'].dt.day
df['journey_month'] = df['Date_of_Journey'].dt.month
df['journey_year'] = df['Date_of_Journey'].dt.year
```

```python
df.head()
```

```python
# NOW we have the seperated features so, we dont need of Date_of_journey column 
```

```python
df.drop('Date_of_Journey',axis = 1,inplace = True)
```

```python
# updating departure and the arrival time
```

```python
def dep_hour(data_set,col):
    df[col+'_hour'] = data_set[col].dt.hour   # here it takes the entire column not the one single cell d
def dep_min(data_set,col):
    df[col+'_minute'] = data_set[col].dt.minute
def drop_col(data_set,col):
    df.drop(col,axis= 1, inplace = True)
```

```python
dep_hour(df,'Dep_Time')
dep_min(df,'Dep_Time')
drop_col(df,'Dep_Time')

dep_hour(df,'Arrival_Time')
dep_min(df,'Arrival_Time')
drop_col(df,'Arrival_Time')
```

```python
df.head()
```

```python
# now we are going to updating the duration column
```

```python
# so first convert that into  a list format to work on it

duration = list(df['Duration'])
```

```python
for i in range(len(duration)):
    if len(duration[i].split(' ')) == 2:
        pass
    else :
        if 'h' in duration[i] :
            duration[i] = duration[i] + ' 0m'
        else :
            duration[i] = '0h ' + duration[i]
        
```

```python
df['Duration'] = duration
```

```python
df.head()
```

```python
# now make that duration into numerical 

def hour(x):                 
    return x.split(' ')[0][0:-1]
def minute(x):
    return x.split(' ')[1][0:-1]
    
```

```python
df['hour_duration'] = df['Duration'].apply(hour)  # to convert series 
df['minute_duration'] = df['Duration'].apply(minute)
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace = True)
```

```python
df.head()
```

```python
df.dtypes    # the date and time are all converted into int type now
```

```python
df['hour_duration'] = df['hour_duration'].astype(int)  # to convert in series type
df['minute_duration'] = df['minute_duration'].astype(int)
```
